```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import welch, windows
fs = 2000
duration=0.2
t = np.arange(0,duration,1/fs)
signal1 = np.sin(2 * np.pi * 400 * t)
signal2 = np.sin(2 * np.pi * 600 * t)
signal3 = signal1 + signal2

plt.figure(figsize=(13, 6))

plt.subplot(3,2,1)
plt.title('Signal 3')
plt.xlabel('Time[second]')
plt.ylabel('Amplitude')
plt.grid(True)
plt.plot(t, signal3)
plt.subplot(3,2,3)
plt.title('Signal 3')
plt.xlabel('Time[second]')
plt.ylabel('Amplitude')
plt.grid(True)
plt.plot(t, signal3)
plt.subplot(3,2,5)
plt.title('Signal 3')
plt.xlabel('Time[second]')
plt.ylabel('Amplitude')
plt.grid(True)
plt.plot(t, signal3)

power=signal3**2
signal3_avg_power=np.mean(power)
signal3_avg_power_db=10*np.log10(signal3_avg_power)
mean_noise=0

snr_db1=-5
noise_db1=signal3_avg_power_db - snr_db1
noise_watts1=10**(noise_db1/10)
noise1=np.random.normal(mean_noise,np.sqrt(noise_watts1),size=len(signal3))
noise_signal1=signal3 + noise1
plt.subplot(3,2,2)
plt.plot(t,noise_signal1)
plt.plot(t,signal3)
plt.title('Signal3 with WSS Noise=[-5dB]')
plt.xlabel('Time[second]')
plt.ylabel('Magnitude')
plt.grid(True)

snr_db2=0
noise_db2=signal3_avg_power_db - snr_db2
noise_watts2=10**(noise_db2/10)
noise2=np.random.normal(mean_noise,np.sqrt(noise_watts2),size=len(signal3))
noise_signal2=signal3 + noise2
plt.subplot(3,2,4)
plt.plot(t,noise_signal2)
plt.plot(t,signal3)
plt.title('Signal3 with WSS Noise=[0dB]')
plt.xlabel('Time[second]')
plt.ylabel('Magnitude')
plt.grid(True)

snr_db3=5
noise_db3=signal3_avg_power_db - snr_db3
noise_watts3=10**(noise_db3/10)
noise3=np.random.normal(mean_noise,np.sqrt(noise_watts3),size=len(signal3))
noise_signal3=signal3 + noise3
plt.subplot(3,2,6)
plt.plot(t,noise_signal3)
plt.plot(t,signal3)
plt.title('Signal3 with WSS Noise=[5dB]')
plt.xlabel('Time[second]')
plt.ylabel('Magnitude')
plt.grid(True)

plt.tight_layout()
plt.show()


```
