```python
import numpy as np
import matplotlib.pyplot as plt
fs = 2000
duration=0.2
t = np.arange(0,duration,1/fs)
signal1 = np.sin(2 * np.pi * 400 * t)
signal2 = np.sin(2 * np.pi * 600 * t)
signal3 = signal1 + signal2

plt.figure(figsize=(17, 10))

plt.subplot(6, 1, 1)
plt.plot(signal1,c='r')
plt.xlabel("No of samples")
plt.ylabel("Amplitude")
plt.title('Signal 1')
plt.grid(True)

plt.subplot(6,1,2)
plt.plot(signal2,c='g')
plt.xlabel("No of samples")
plt.ylabel("Amplitude")
plt.title('Signal 2')
plt.grid(True)

plt.subplot(6,1,3)
plt.plot(t, signal3)
plt.title('Signal 3')
plt.xlabel('Time[second]')
plt.ylabel('Amplitude')
plt.grid(True)

fft_result = np.fft.fft(signal3)
freq_values = np.fft.fftfreq(len(signal3), 1/fs)
plt.subplot(6,1,4)
plt.plot(freq_values,fft_result,c='m')
plt.xlabel('Frequency in Hz')
plt.ylabel('Amplitude')
plt.title('FFT of signal3')
magnitude_spectrum = np.abs(fft_result)
phase_spectrum = np.angle(fft_result)
plt.grid(True)

plt.subplot(6,1,5)
plt.plot(freq_values,magnitude_spectrum,c='y')
plt.title('Magnitude Spectrum of FFT')
plt.xlabel('Frequency in Hz')
plt.ylabel('Magnitude')
plt.grid(True)

plt.subplot(6,1,6)
plt.plot(freq_values,phase_spectrum,c='c')
plt.title('Phase Spectrum of FFT')
plt.xlabel('Frequency in Hz')
plt.ylabel('Phase [radians]')
plt.grid(True)

plt.tight_layout()
plt.show()

```
