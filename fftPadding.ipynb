```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import stft
from scipy.signal import windows
fs = 2000
duration=0.2
t = np.arange(0,duration,1/fs)
signal1 = np.sin(2 * np.pi * 400 * t)
signal2 = np.sin(2 * np.pi * 600 * t)
signal3 = signal1 + signal2

plt.figure(figsize=(17, 10))

padded_signal3 = np.pad(signal3, (0, 768 - len(signal3)), 'constant')
signal3_truncated1 =padded_signal3[:256]

print(f"Total samples in signal3_truncated1: {len(signal3_truncated1)}")
fft_result1 = np.fft.fft(signal3_truncated1)
freq_values1 = np.fft.fftfreq(len(signal3_truncated1), 1/fs)
plt.subplot(5,2,1)
plt.plot(freq_values1,fft_result1)
plt.xlabel('Frequency in Hz')
plt.ylabel('Amplitude')
plt.title('FFT of signal3_truncated1[0-256]')
magnitude_spectrum1 = np.abs(fft_result1)
plt.subplot(5,2,2)
plt.plot(freq_values1,magnitude_spectrum1)
plt.title('Magnitude Spectrum of FFT')
plt.xlabel('Frequency in Hz')
plt.ylabel('Magnitude')
plt.grid(True)


signal3_truncated2 =padded_signal3[128:384]

print(f"Total samples in signal3_truncated2: {len(signal3_truncated2)}")
fft_result2 = np.fft.fft(signal3_truncated2)
freq_values2 = np.fft.fftfreq(len(signal3_truncated2), 1/fs)
plt.subplot(5,2,3)
plt.plot(freq_values2,fft_result2)
plt.xlabel('Frequency in Hz')
plt.ylabel('Amplitude')
plt.title('FFT of signal3_truncated2[128-384]')
magnitude_spectrum2 = np.abs(fft_result2)
plt.subplot(5,2,4)
plt.plot(freq_values2,magnitude_spectrum2)
plt.title('Magnitude Spectrum of FFT')
plt.xlabel('Frequency in Hz')
plt.ylabel('Magnitude')
plt.grid(True)


signal3_truncated3 =padded_signal3[256:512]

print(f"Total samples in signal3_truncated3: {len(signal3_truncated3)}")
fft_result3 = np.fft.fft(signal3_truncated3)
freq_values3 = np.fft.fftfreq(len(signal3_truncated3), 1/fs)
plt.subplot(5,2,5)
plt.plot(freq_values3,fft_result3)
plt.xlabel('Frequency in Hz')
plt.ylabel('Amplitude')
plt.title('FFT of signal3_truncated3[256-512]')
magnitude_spectrum3 = np.abs(fft_result3)
plt.subplot(5,2,6)
plt.plot(freq_values3,magnitude_spectrum3)
plt.title('Magnitude Spectrum of FFT')
plt.xlabel('Frequency in Hz')
plt.ylabel('Magnitude')
plt.grid(True)

signal3_truncated4 =padded_signal3[384:640]

print(f"Total samples in signal3_truncated4: {len(signal3_truncated4)}")
fft_result4 = np.fft.fft(signal3_truncated4)
freq_values4 = np.fft.fftfreq(len(signal3_truncated4), 1/fs)
plt.subplot(5,2,7)
plt.plot(freq_values4,fft_result4)
plt.xlabel('Frequency in Hz')
plt.ylabel('Amplitude')
plt.title('FFT of signal3_truncated4[384-640]')
magnitude_spectrum4 = np.abs(fft_result4)
plt.subplot(5,2,8)
plt.plot(freq_values4,magnitude_spectrum4)
plt.title('Magnitude Spectrum of FFT')
plt.xlabel('Frequency in Hz')
plt.ylabel('Magnitude')
plt.grid(True)

signal3_truncated5 =padded_signal3[512:768]

print(f"Total samples in signal3_truncated5: {len(signal3_truncated5)}")
fft_result5 = np.fft.fft(signal3_truncated5)
freq_values5 = np.fft.fftfreq(len(signal3_truncated5), 1/fs)
plt.subplot(5,2,9)
plt.plot(freq_values5,fft_result5)
plt.xlabel('Frequency in Hz')
plt.ylabel('Amplitude')
plt.title('FFT of signal3_truncated5[512-768]')
magnitude_spectrum5 = np.abs(fft_result5)
plt.subplot(5,2,10)
plt.plot(freq_values5,magnitude_spectrum5)
plt.title('Magnitude Spectrum of FFT')
plt.xlabel('Frequency in Hz')
plt.ylabel('Magnitude')
plt.grid(True)

plt.tight_layout()
plt.show()
```
