```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import stft
from scipy.signal import get_window

fs = 2000
duration=0.2
t = np.arange(0,duration,1/fs)
signal1 = np.sin(2 * np.pi * 400 * t)
signal2 = np.sin(2 * np.pi * 600 * t)
signal3 = signal1 + signal2


plt.figure(figsize=(17, 10))

total_length = 768
padding_amount = total_length - len(signal3)
padding_each_side = padding_amount // 2

# Apply symmetric padding
padded_signal3 = np.pad(signal3, (padding_each_side, padding_amount - padding_each_side), 'constant')
signal3_truncated1 =padded_signal3[:256]
window1 = get_window('hann', len(signal3_truncated1))
windowed_signal1 = signal3_truncated1 * window1

print(f"Total samples in signal3_truncated1: {len(signal3_truncated1)}")
fft_result1 = np.fft.fft(windowed_signal1)
freq_values1 = np.fft.fftfreq(len(windowed_signal1), 1/fs)
plt.subplot(5,2,1)
plt.plot(freq_values1,fft_result1)
plt.xlabel('Frequency in Hz')
plt.ylabel('Amplitude')
plt.title('FFT of signal3_truncated1[0-256]')
magnitude_spectrum1 = np.abs(fft_result1)
magnitude_spectrum1_dB = 20 * np.log10(magnitude_spectrum1)
plt.subplot(5,2,2)
plt.plot(freq_values1,magnitude_spectrum1_dB )
plt.title('Magnitude Spectrum of FFT[dB]')
plt.xlabel('Frequency in Hz')
plt.ylabel('Magnitude')
plt.ylim(-50, 40)
plt.grid(True)


signal3_truncated2 =padded_signal3[128:384]
window2 = get_window('hann', len(signal3_truncated2))
windowed_signal2 = signal3_truncated2 * window2

print(f"Total samples in signal3_truncated2: {len(signal3_truncated2)}")
fft_result2 = np.fft.fft(windowed_signal2)
freq_values2 = np.fft.fftfreq(len(windowed_signal2), 1/fs)
plt.subplot(5,2,3)
plt.plot(freq_values2,fft_result2)
plt.xlabel('Frequency in Hz')
plt.ylabel('Amplitude')
plt.title('FFT of signal3_truncated2[128-384]')
magnitude_spectrum2 = np.abs(fft_result2)
magnitude_spectrum2_dB = 20 * np.log10(magnitude_spectrum2)
plt.subplot(5,2,4)
plt.plot(freq_values2,magnitude_spectrum2_dB)
plt.title('Magnitude Spectrum of FFT[dB]')
plt.xlabel('Frequency in Hz')
plt.ylabel('Magnitude')
plt.ylim(-50, 40)
plt.grid(True)


signal3_truncated3 =padded_signal3[256:512]
window3 = get_window('hann', len(signal3_truncated3))
windowed_signal3 = signal3_truncated3 * window3

print(f"Total samples in signal3_truncated3: {len(signal3_truncated3)}")
fft_result3 = np.fft.fft(windowed_signal3 )
freq_values3 = np.fft.fftfreq(len(windowed_signal3 ), 1/fs)
plt.subplot(5,2,5)
plt.plot(freq_values3,fft_result3)
plt.xlabel('Frequency in Hz')
plt.ylabel('Amplitude')
plt.title('FFT of signal3_truncated3[256-512]')
magnitude_spectrum3 = np.abs(fft_result3)
magnitude_spectrum3_dB = 20 * np.log10(magnitude_spectrum3)
plt.subplot(5,2,6)
plt.plot(freq_values3,magnitude_spectrum3_dB)
plt.title('Magnitude Spectrum of FFT[dB]')
plt.xlabel('Frequency in Hz')
plt.ylabel('Magnitude')
plt.ylim(-50, 40)
plt.grid(True)

signal3_truncated4 =padded_signal3[384:640]
window4 = get_window('hann', len(signal3_truncated4))
windowed_signal4 = signal3_truncated4 * window4

print(f"Total samples in signal3_truncated4: {len(signal3_truncated4)}")
fft_result4 = np.fft.fft(windowed_signal4)
freq_values4 = np.fft.fftfreq(len(windowed_signal4), 1/fs)
plt.subplot(5,2,7)
plt.plot(freq_values4,fft_result4)
plt.xlabel('Frequency in Hz')
plt.ylabel('Amplitude')
plt.title('FFT of signal3_truncated4[384-640]')
magnitude_spectrum4 = np.abs(fft_result4)
magnitude_spectrum4_dB = 20 * np.log10(magnitude_spectrum4)
plt.subplot(5,2,8)
plt.plot(freq_values4,magnitude_spectrum4_dB)
plt.title('Magnitude Spectrum of FFT[dB]')
plt.xlabel('Frequency in Hz')
plt.ylabel('Magnitude')
plt.ylim(-50, 40)
plt.grid(True)

signal3_truncated5 =padded_signal3[512:768]
window5 = get_window('hann', len(signal3_truncated5))
windowed_signal5 = signal3_truncated5 * window5

print(f"Total samples in signal3_truncated5: {len(signal3_truncated5)}")
fft_result5 = np.fft.fft(windowed_signal5)
freq_values5 = np.fft.fftfreq(len(windowed_signal5), 1/fs)
plt.subplot(5,2,9)
plt.plot(freq_values5,fft_result5)
plt.xlabel('Frequency in Hz')
plt.ylabel('Amplitude')
plt.title('FFT of signal3_truncated5[512-768]')
magnitude_spectrum5 = np.abs(fft_result5)
magnitude_spectrum5_dB = 20 * np.log10(magnitude_spectrum5)
plt.subplot(5,2,10)
plt.plot(freq_values5,magnitude_spectrum5_dB)
plt.title('Magnitude Spectrum of FFT[dB]')
plt.xlabel('Frequency in Hz')
plt.ylabel('Magnitude')
plt.ylim(-50, 40)
plt.grid(True)

plt.tight_layout()
plt.show()
```
